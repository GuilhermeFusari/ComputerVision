```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

 <h1> Carregando a imagem</h1>


```python
imagem = cv2.imread('miu.jpg')
```

```python
type(imagem)
```

<h1>Plotando a img</h1>


```python
imagem.shape
```

```python
plt.imshow(imagem) #opencv2 Trabalha com BGR ao contrario do RGB do matplotlib
```

```python
#correo de cor da img
img_cor = cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cor)
```

<h1>Redimensionando a img</h1>


```python
img_redimensionada = cv2.resize(img_cor,(640, 480))
```

```python
plt.imshow(img_redimensionada)
```

```python
#outra maneira de redimensionar sem distorcer 
#neste caso 50% do tamanho dela
largura = 0.5
altura = 0.5
img_nova = cv2.resize(img_cor,(0,0),img_cor,largura,altura)
```

```python
plt.imshow(img_nova)
```

<h1>Salvar imgs</h1>

```python
cv2.imwrite('miu3.jpg',img_nova)
```
